```python
#importing neccessary libraries
import time
import numpy as np
from scipy import stats
import networkx as nx
from tabulate import tabulate
import matplotlib.pyplot as plt
import networkx.algorithms.community as nx_comm
from sklearn.cluster import SpectralClustering
%matplotlib inline
```

```python
""" Reading the follwoing Graphs """
karate = nx.read_gml('/content/gdrive/MyDrive/Sem-3/Social media analysis/Assignment-1/karate.gml', label = 'id')
dolphins = nx.read_gml('/content/gdrive/MyDrive/Sem-3/Social media analysis/Assignment-1/dolphins.gml')
jazz = nx.read_weighted_edgelist('/content/gdrive/MyDrive/Sem-3/Social media analysis/Assignment-1/jazz.net')
```

```python
"""Statistics:
   Computing the following statistics describing the datasets:
    • number of nodes n,
    • number of edges m
    • average path lengths d ̄
    • average clustering coefficient C"""

statistics = [["Social Network", "Number of Nodes", "Number of Edges", "Average Path Lengths", "Average Clustering Cofficient"],
         ["Karate", len(karate.nodes), len(karate.edges), nx.average_shortest_path_length(karate), nx.average_clustering(karate)],
         ["Dolphins", len(dolphins.nodes), len(dolphins.edges), nx.average_shortest_path_length(dolphins), nx.average_clustering(dolphins)],
         ["Jazz", len(jazz.nodes), len(jazz.edges), nx.average_shortest_path_length(jazz), nx.average_clustering(jazz)]]

print(tabulate(statistics))
```

```python
""" Running Spectral Clustering
    On KARATE Dataset """

adj_matrix = nx.to_numpy_matrix(karate)

start = time.time()
scKarate = SpectralClustering(n_clusters=4, affinity='precomputed', n_init=100)
# find communities in the graph
scKarate.fit(adj_matrix)
# find the nodes forming the communities
labels = scKarate.labels_

karateClusters = [[] for i in range(4)]
i = 1
for x in labels:
    if(x == 0):
        karateClusters[0].append(i)
        i += 1
    elif(x == 1):
        karateClusters[1].append(i)
        i += 1
    elif(x == 2):
        karateClusters[2].append(i)
        i += 1
    elif(x == 3):
        karateClusters[3].append(i)
        i += 1
        
end = time.time()
karateTimeTaken = end - start
```

```python
""" Running Spectral Clustering
    On DOLPHINS Dataset """

adj_matrix = nx.to_numpy_matrix(dolphins)

start = time.time()
scDolphins = SpectralClustering(n_clusters=4, affinity='precomputed', n_init=100)
# find communities in the graph
scDolphins.fit(adj_matrix)
# find the nodes forming the communities
labels = scDolphins.labels_

dolphinsClusters = [[] for i in range(4)]
i = 1
for x in labels:
    if(x == 0):
        dolphinsClusters[0].append(i)
        i += 1
    elif(x == 1):
        dolphinsClusters[1].append(i)
        i += 1
    elif(x == 2):
        dolphinsClusters[2].append(i)
        i += 1
    elif(x == 3):
        dolphinsClusters[3].append(i)
        i += 1
        
end = time.time()
dolphinsTimeTaken = end - start
```

```python
""" Running Spectral Clustering
    On JAZZ Dataset """

adj_matrix = nx.to_numpy_matrix(jazz)

start = time.time()
scJazz = SpectralClustering(n_clusters=4, affinity='precomputed', n_init=100)
# find communities in the graph
scJazz.fit(adj_matrix)
# find the nodes forming the communities
labels = scJazz.labels_

jazzClusters = [[] for i in range(4)]
i = 1
for x in labels:
    if(x == 0):
        jazzClusters[0].append(i)
        i += 1
    elif(x == 1):
        jazzClusters[1].append(i)
        i += 1
    elif(x == 2):
        jazzClusters[2].append(i)
        i += 1
    elif(x == 3):
        jazzClusters[3].append(i)
        i += 1
        
end = time.time()
jazzTimeTaken = end - start
```

```python
table = [["Social Network", "Number of clusters found", "Modularity score for this clustering", "Run time"],
         ["Karate", len(karateClusters), str(nx_comm.modularity(karate, karateClusters)), karateTimeTaken],
         ["Dolphins", len(dolphinsClusters), 0, dolphinsTimeTaken],
         ["Jazz", len(jazzClusters), 0, jazzTimeTaken]]

print(tabulate(table))
```
